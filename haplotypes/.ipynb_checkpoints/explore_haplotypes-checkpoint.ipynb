### This script can be used to play around and visualise the haplotypes for a given genomic window

```python
# Set the population to use for this analysis
study_pop = 'Madina_gambiae_Delta'
# And the genomic region
region = '2R:28401363-28416200'
```

```python
ag3 = malariagen_data.Ag3('gs://vo_agam_release', pre = True)
ag3.sample_sets(release = '3.2')
ag3
```

```python
# Subset to only the phenotypes that we are interested in
meta['population'] = meta[['location', 'species', 'insecticide']].apply('_'.join, 1)
meta.query(f'population == "{study_pop}"', inplace = True)
focal_samples = meta.index
```

```python
# Get the haplotypes from the region of interest
haps = ag3.haplotypes(
    region = region,
    analysis = 'gamb_colu', 
    sample_sets = '3.2',
    sample_query = f'sample_id in {list(meta.sample_id)}',
)
haps
```

```python
haparray = allel.GenotypeArray(haps['call_genotype']).to_haplotypes()
hapsamples = np.array(np.repeat(haps['sample_id'], 2))
```

```python
dist = allel.pairwise_distance(haparray, metric = 'hamming')
dist
```

```python
site_filter = ag3.snp_calls(region=region, sample_sets="3.2")['variant_filter_pass_gamb_colu']
n_bases = np.sum(site_filter.values)
dist_dxy = dist * haparray.n_variants / n_bases
```

```python

def find_clusters(dist, n, threshold=0.001, method='complete'):

        # build hierarchy
        clust = scipy.cluster.hierarchy.linkage(dist, method=method)

        # find clusters
        f = scipy.cluster.hierarchy.fcluster(clust, threshold,
                                             criterion='distance')

        # compute cluster sizes
        fsz = np.bincount(f)

        # sort largest first
        fsort = np.argsort(fsz)[::-1]

        # take largest n
        fsort = fsort[:n]

        # get haplotype indices for each cluster
        clusters = [set(np.nonzero(f == i)[0]) for i in fsort]

        return clusters

focal_clusters = find_clusters(dist_dxy, n=50, threshold=0.001)
```

```python
min_cluster_size = 20
large_focal_clusters = [cluster for cluster in focal_clusters if len(cluster) > min_cluster_size]
print(f'There were {len(large_focal_clusters)} clusters larger than {min_cluster_size} haplotypes.')
```

```python
def assign_sample_haplotypes(cluster_ids):
    cluster_ids = focal_clusters[0]
    cluster_samples = hapsamples[list(cluster_ids)]
    sample_hap_genotype = meta.sample_id.apply(lambda x: np.sum(np.isin(cluster_samples, x)))
    return(sample_hap_genotype)

for i in range(len(large_focal_clusters)):
    meta[f'cluster_{i+1}'] = assign_sample_haplotypes(large_focal_clusters[i])

output_table = meta[['phenotype'] + [f'cluster_{i+1}' for i in range(len(large_focal_clusters))]]
output_table
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


def load_haplotype_pdist_dxy(haparray, site_filter, pos, start, stop):

    #h = allel.GenotypeArray(
        #ag3.haplotypes(
            #region = chrom + ':' + str(start) + '-' + str(stop),
            #analysis = 'gamb_colu', 
            #sample_sets = '3.2',
            #sample_query = f'sample_id in {list(meta.sample_id)}',
        #)['call_genotype']
    #).to_haplotypes()

    pos_locate = pos.locate_range(start, stop)
    h = haparray[pos_locate, :]
    # compute pairwise Hamming distance
    dist = allel.pairwise_distance(h, metric='hamming')

    # convert to Dxy distance
    is_accessible = site_filter[pos_locate]
    n_bases = np.count_nonzero(is_accessible)
    dist_dxy = dist * h.n_variants / n_bases

    return dist_dxy

def load_moving_haplotype_pdist_dxy(haparray, site_filter, positions, size,
                                    rebuild=False):

    # compute windows
    pos = allel.SortedIndex(positions)
    windows = allel.moving_statistic(pos,
                                     statistic=lambda v: [v[0], v[-1]], size=size)

    dscan = []
    for window_start, window_stop in windows:
        dist = load_haplotype_pdist_dxy(haparray, site_filter, pos, window_start, window_stop)
        dscan.append(dist)

    return windows, dscan

def plot_ibs_clusters(chrom, start, stop, size,
                      start_index=0, stop_index=None, merge=1,
                      threshold=0.001, n_clusters=3, min_clen=5):

    region = chrom + ':' + str(start) + '-' + str(stop)
    haps = ag3.haplotypes(
        region = region,
        analysis = 'gamb_colu', 
        sample_sets = '3.2',
        sample_query = f'sample_id in {list(meta.sample_id)}',
    )
    
    haparray = allel.GenotypeArray(haps['call_genotype']).to_haplotypes()
    
    positions = haps['variant_position'].values
    
    site_filter = ag3.snp_calls(
        region=region, 
        sample_sets="3.2"
    )['variant_filter_pass_gamb_colu']
        
    windows, dscan = load_moving_haplotype_pdist_dxy(haparray, site_filter, positions, size)

    windows = windows[start_index:stop_index]
    dscan = dscan[start_index:stop_index]

    fig = plt.figure(figsize=(12, 4))

    ax = plt.subplot2grid((6, 1), (0, 0), rowspan=5)
    sns.despine(ax=ax, offset=5, bottom=True)
    #ax.set_xticks([])

    previous_x = None
    previous_clusters = None
    previous_clens = None

    for (start_pos, stop_pos), dist in zip(windows, dscan):

        # determine X coord
        x = np.mean([start_pos, stop_pos])

        # find clusters
        clusters = find_clusters(dist, n_clusters, threshold=threshold)
        clusters = [c for c in clusters if len(c) > min_clen]

        # compute cluster sizes
        clens = [len(cluster) for cluster in clusters]

        # plot cluster sizes
        ax.scatter([x] * len(clusters), clens, c='w')

        # plot lines connecting clusters
        if previous_clusters:
            for cluster in clusters:
                for previous_cluster in previous_clusters:
                    n_common = len(cluster.intersection(previous_cluster))
                    if n_common > 0:
                        l = plt.Line2D([previous_x, x], [len(previous_cluster), len(cluster)],
                                       color='k', lw=0.1*n_common, alpha=1, zorder=-10)
                        ax.add_artist(l)

        previous_x = x
        previous_clusters = clusters
        previous_clens = clens

    ax.set_ylabel('cluster size')
    ax.set_ylim(bottom=0)

   # ax = plt.subplot2grid((6, 1), (5, 0))
   # sns.despine(ax=ax, offset=5)
   # start, stop = windows[0, 0], windows[-1, 1]
   # self.hap_data.plot_genes(chrom, start, stop, ax=ax, label=True, label_unnamed=False, height=.1)
    ax.set_xlabel('%s position (bp)' % chrom)

```

```python
plot_ibs_clusters('2R', 28380000, 28440000, 1000, n_clusters = 5)
```
